```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Y:
    
    """
    The vertical motion of a ball
    """
    
    def __init__(self,v0):
        self.v0 = v0
        self.g = 9.81
    
    def value(self,t):
        result = self.v0*t - 0.5*self.g*t**2
        return result
    
    def formula(self):
        return 'v0*t - 0.5*g*t**2; v0=%g' % self.v0
    
    def graph(self, a, b):
        t = np.arange(a,b,0.01)
        plt.plot(t, self.value(t))
        
```

```python
y = Y(3)
```

```python
v=y.value(0.1)
print(v)
```

```python
y = Y(5)
t = 0.2
v = y.value(t)

print(f'y(t={t}; v0={y.v0}) = {v}')
print(y.formula())
```

```python
for k in range(0,7):
    y = Y(k*30)
    y.graph(-20,20)
    
plt.title('So Pretty!')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
class SelfExplorer:
    def __init__(self, a):
        self.a = a
        print('init: a=%g, id(self)=%d' % (self.a, id(self)))
    def value(self, x):
        print('value: a=%g, id(self)=%d' % (self.a, id(self)))
        return self.a*x
```

```python
s1 = SelfExplorer(1)
s2 = SelfExplorer(2)
```

```python
class V:
    
    """
    Velocity profile for pipeflow
    - The velocity v varies through the pipe cross section
    - R is the radius of the pipe
    - beta is the pressure gradient (the force that drives the flow through the pipe)
    - mu0 is a viscosity coefficient (small for air, larger for water 
    and even larger for toothpaste)
    - n is a real number reflecting the viscous properties
    of the fluid (n = 1 for water and air, n < 1 for many modern plastic materials)
    - r is a radial coordinate that measures the distance from the centerline (r = 0 is the
    centerline, r = R is the pipe wall).
    """
    
    def __init__(self, beta, mu0, n, R):
        self.beta, self.mu0, self.n, self.R = beta, mu0, n , R
        
    def value(self,r):
        beta, mu0, n , R = self.beta, self.mu0, self.n, self.R

        n = float(n) # ensure float division
        
        v = (beta/(2*mu0))**(1/n)*(n/(n+1))*(R**(1+(1/n))-r**(1+(1/n)))
        
        return v
        
    def graph(self):
        r=np.arange(0,self.R,0.0001)
        #normalized curve
        curve = self.value(r)/self.value(0)
        plt.plot(r,curve)
        
```

```python
v = V(3,0.02,0.1,0.02)
v.graph()
```

```python
ns = np.flip(np.arange(0.1,1,0.1))

for n in ns:
    v = V(3,0.02,n,0.1)
    v.graph()

plt.title('Velocity profile for pipeflow at different viscosity')
plt.xlabel('Radial Coordinate')
plt.ylabel('Velocity')
plt.show()
```
